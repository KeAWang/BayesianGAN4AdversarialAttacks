```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
%matplotlib inline
```

```python
data = []
for i in range(10,160,10):
    with open("results_"+str(i)+".json") as f:
        res = json.load(f)
        data.append(res)
        
data = dict(zip(data[0],zip(*[d.values() for d in data])))
```

```python
data.keys()
```

```python
x = range(10,160,10)
plt.plot(x,data['adversarial_uncertainty_correct'],'o-',label='Defeated Adversary')
plt.plot(x,data['adversarial_uncertainty_incorrect'],'o-', label='Lost to Adversary')
plt.xlabel('Iteration')
plt.ylabel('Classification confidence')
plt.legend()
```
