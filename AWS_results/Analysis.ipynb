```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
%matplotlib inline
```

```python
data = []
for i in range(100,4100,100):
    with open("results_"+str(i)+".json") as f:
        res = json.load(f)
        data.append(res)
        
data = dict(zip(data[0],zip(*[d.values() for d in data])))
```

```python
data.keys()
```

```python
for k in data.keys():
    plt.figure()
    plt.plot(data[k])
    plt.title(k)
    plt.show
```
