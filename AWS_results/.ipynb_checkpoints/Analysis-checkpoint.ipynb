```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
%matplotlib inline
```

```python
data = []
for i in range(100,4100,100):
    with open("results_"+str(i)+".json") as f:
        res = json.load(f)
        data.append(res)
        
data = dict(zip(data[0],zip(*[d.values() for d in data])))
```

```python
data.keys()
```

```python
data['previous_chkpt']
```

```python
f, axarr = plt.subplots(2, sharex=True, figsize=(8,8))
#f.suptitle('Sharing X axis')
x = range(1,41)
axarr[0].plot(x,1-np.array(data['adversarial_classification_accuracy']), '-o')
axarr[0].set_title('Adversarial accuracy on adversarial inputs')
axarr[0].set(ylabel='Accuracy')
axarr[1].plot(x,data['non_adversarial_classification_accuracy'], '-o')
axarr[1].set_title('Adversarial accuracy on real inputs')
axarr[1].set(xlabel='Epoch', ylabel='Accuracy', xlim=(0,40))

#Classified not fake/total labelled test on adversarial input
```

```python
plt.plot(x,0.5*(np.array(data['non_adversarial_classification_accuracy'])+ 1-np.array(data['adversarial_classification_accuracy'])), '-o')
plt.title('Adversarial accuracy assuming equal distribution')
```

```python
plt.plot(data['adversarial_unfilted_semi_supervised_acc'], '-o')
plt.title('MNIST classification accuracy on adversarial input')
```

```python
x = range(1,41)
plt.plot(x, data['semi_supervised_acc'],'o-')
plt.title('MNIST Semi-supervised training')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
```
